```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def lashuk(l, n): return np.sqrt(l) * (n/l)**(2/3) 
def ibeid(l, n): return np.log(l) * (n/l)**(2/3)
```

```python
l = np.arange(100, 1000)
n = l*1e6


plt.plot(lashuk(l, n), label='Lashuk')
plt.plot(ibeid(l, n), label='Ibeid')

plt.title(f"(N/L) = {1e6}")
plt.legend()
plt.savefig('lashuk_vs_ibeid', dpi=300)
```
