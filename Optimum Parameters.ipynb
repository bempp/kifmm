```python
import pandas as pd
```

```python
f32_blas_1 = pd.read_csv('grid_search_laplace_f32_1_blas_amd.csv')
f32_fft_1 = pd.read_csv('grid_search_laplace_f32_1_fft_amd.csv')
f32_blas_8 = pd.read_csv('grid_search_laplace_f32_8_blas_amd.csv')
f32_fft_8 = pd.read_csv('grid_search_laplace_f32_8_fft_amd.csv')

f64_blas_1 = pd.read_csv('grid_search_laplace_f64_1_blas_amd.csv')
f64_fft_1 = pd.read_csv('grid_search_laplace_f64_1_fft_amd.csv')
f64_blas_8 = pd.read_csv('grid_search_laplace_f64_8_blas_amd.csv')
f64_fft_8 = pd.read_csv('grid_search_laplace_f64_8_fft_amd.csv')

data_amd = {
    "f32": 
    {
        "1": {"blas": f32_blas_1, "fft": f32_fft_1},
        "8": {"blas": f32_blas_8, "fft": f32_fft_8},
    },
    "f64":
    {
        "1": {"blas": f64_blas_1, "fft": f64_fft_1},
        "8": {"blas": f64_blas_8, "fft": f64_fft_8},
    }
    
}
```

```python
def best_parameters(df, threshold):

    tmp = df[ (df['rel_l2_err'] >= 0.1 * threshold) & (df['rel_l2_err'] <= threshold)]

    # return tmp
    # Filter for row with best runtime
    idx = tmp['runtime'].idxmin()
    return tmp.loc[idx]
```

```python
threshold_f32 = [1e-3, 1e-4]
threshold_f64 = [1e-6, 1e-8, 1e-10]
```

```python
for threshold in threshold_f32:
    print(f"m2l: blas, threshold: {threshold} n_points: 1e6")
    print(best_parameters(f32_blas_1, threshold)[["runtime", 'depth', 'surface_diff', 'svd_threshold', 'expansion_order', 'n_oversamples']])
    print()

for threshold in threshold_f32:
    print(f"m2l: fft, threshold: {threshold}, n_points: 1e6")
    print(best_parameters(f32_fft_1, threshold)[["runtime", 'depth', 'expansion_order', 'block_size']])
    print()

print('-'*50)

for threshold in threshold_f64:
    print(f"m2l: blas, threshold: {threshold} n_points: 1e6")
    print(best_parameters(f64_blas_1, threshold)[["runtime", 'depth', 'surface_diff', 'svd_threshold', 'expansion_order', 'n_oversamples']])
    print()

for threshold in threshold_f64:
    print(f"m2l: fft, threshold: {threshold}, n_points: 1e6")
    print(best_parameters(f64_fft_1, threshold)[["runtime", 'depth', 'expansion_order', 'block_size']])
    print()
```
